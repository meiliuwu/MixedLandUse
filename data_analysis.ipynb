## CBG-based stats Batch Process

```python
import pandas as pd
import numpy as np
import os
import geopandas as gpd
import matplotlib.pyplot as plt
import datetime
from collections import Counter
import json 

path='/media/raid/SafeGraph/nyc_park/los_angeles/'
output=None
files=os.listdir(path)
files.sort()
for filename in files:
    if '.csv' in filename:
        cbg_visitors = Counter()
        date=str(filename.split('.')[0].split('_')[2])
        print(date)
        df_join=pd.read_csv(path+'/'+filename)
        for index, row in df_join.iterrows():
            visitor_home_cbgs=json.loads(row['visitor_home_cbgs'])
            cbg_visitors.update(Counter(visitor_home_cbgs))
        #print(len(dict(cbg_visitors).keys()))
        df_visitors=pd.DataFrame.from_dict(dict(cbg_visitors),orient='index')
        df_visitors['cbg']=df_visitors.index
        df_visitors.reset_index(drop=True,inplace=True)
        #df_visitors.index.name='cbg'
        mapping = {df_visitors.columns[0]:date}
        df_visitors = df_visitors.rename(columns=mapping)
        print(df_visitors.head(5))
        if output is None:
            output=df_visitors
            output.reset_index(inplace=True, drop=True) 
        else: 
            output = pd.concat([output, df_visitors], axis=1, join='inner')
            #output==output.merge(df_visitors, how='inner', on='cbg', left_index=False, right_index=False)
output = output.loc[:,~output.columns.duplicated()] #remove duplicate columns
output.set_index('cbg',inplace=True)
output.to_csv('./los_angeles/cbg_la_visitors_weekly.csv',index=True)
output
```

```python
## Create Time Series DataFrame
print(output.sum())
output.sum().plot()
```

```python
## Filter by Demographics and Economic Variables
cbgs_la=list(output.index.unique())
cbgs_la = list(filter(lambda cbg: cbg.startswith('06037'), cbgs))
print(len(cbgs_la))

df_income=pd.read_csv("/media/raid//SafeGraph/data_portal/census/income_cbg.csv",dtype=object)
df_income=df_income[['cbg','household_median_income']]
df_income.dropna(inplace=True)
df_income=df_income.loc[(df_income['cbg'].isin(cbgs_la))] 
df_income

df_pop=pd.read_csv("/media/raid//SafeGraph/data_portal/census/pop_cbg.csv",dtype=object)
df_pop=df_pop[['cbg','pop']]
df_pop.dropna(inplace=True)
df_pop=df_pop.loc[(df_pop['cbg'].isin(cbgs_la))] 
df_pop

df_ethnicity=pd.read_csv("/media/raid//SafeGraph/data_portal/census/ethnicity_cbg.csv",dtype=object)
df_ethnicity['r_hispanic']=pd.to_numeric(df_ethnicity['hispanic'])*100/pd.to_numeric(df_ethnicity['pop'])
df_ethnicity=df_ethnicity[['cbg','r_hispanic']]
df_ethnicity.dropna(inplace=True)
df_ethnicity=df_ethnicity.loc[(df_ethnicity['cbg'].isin(cbgs_la))] 
df_ethnicity

df_race=pd.read_csv("/media/raid//SafeGraph/data_portal/census/race_cbg.csv",dtype=object)
df_race['r_white']=pd.to_numeric(df_race['white'])*100/pd.to_numeric(df_race['pop'])
df_race=df_race[['cbg','r_white']]
df_race.dropna(inplace=True)
df_race=df_race.loc[(df_race['cbg'].isin(cbgs_la))] 
df_race
```

## Median Household Income

```python
import numpy as np
from matplotlib.pyplot import *

df_join['household_median_income']=pd.to_numeric(df_join['household_median_income'])
print(df_join['household_median_income'].min())
print(df_join['household_median_income'].max())
quintiles=np.percentile(df_join.household_median_income, [0, 20, 40, 60, 80, 100])
print(quintiles,quintiles[-2])

df_high=df_join.loc[df_join['household_median_income']>=quintiles[-2]]
df_high=df_high[df_high.columns[~df_high.columns.isin(['household_median_income'])]]
print(df_high.head(5))
df_high.sum().plot()

df_low=df_join.loc[df_join['household_median_income']<=quintiles[1]]
df_low=df_low[df_low.columns[~df_low.columns.isin(['household_median_income'])]]
print(df_low.head(5))

df_low.sum().plot()
legend(['Higher Income Quintile', 'Lower Income Quintile'])
```

```python
## add population weight
df_join_wt=df_join.merge(df_pop,on='cbg')
df_join_wt.set_index('cbg',inplace=True, drop=True) 

for (columnName, columnData) in df_join_wt.iteritems():
    if '-' in columnName:
        df_join_wt[columnName]=pd.to_numeric(columnData.values)*10000.0/pd.to_numeric(df_join_wt['pop'])
df_join_wt
```
